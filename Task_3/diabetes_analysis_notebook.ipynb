```python
import diabetes_library.feature_engineering as fe
import diabetes_library.preprocessing as pr
import diabetes_library.train_model as tr
import pandas as pd
```

```python
# Load and split data
data_path = "sample_diabetes_mellitus_data.csv"
data_loader = pr.DataLoader(data_path)
train_data, test_data = data_loader.load_and_split_data()
```

```python
# Remove NaN
train_nan_remover = pr.NanRemover(train_data)
test_nan_remover = pr.NanRemover(test_data)

train_data = train_nan_remover.remove_nan(['age', 'gender', 'ethnicity'])
test_data = test_nan_remover.remove_nan(['age', 'gender', 'ethnicity'])
```

```python
# Impute means
train_nan_imputer = pr.NanImputer(train_data)
test_nan_imputer = pr.NanImputer(test_data)

train_data = train_nan_imputer.fill_nan(['height', 'weight'])
test_data = test_nan_imputer.fill_nan(['height', 'weight'])
```

```python
# Transform columns in dataframes
height_standardize_train = fe.Standardizer(train_data)
height_standardize_train.transform(['height'])

height_standardize_test = fe.Standardizer(test_data)
height_standardize_test.transform(['height'])

weight_standardize_train = fe.Normalizer(train_data)
weight_standardize_train.transform(['weight'])

weight_standardize_test = fe.Normalizer(test_data)
weight_standardize_test.transform(['weight'])
```

```python
# Train model, predict, report accuracy score
target_cols = ['age',
               'height',
               'weight',
               'aids',
               'cirrhosis',
               'hepatic_failure',
               'immunosuppression',
               'leukemia',
               'lymphoma',
               'solid_tumor_with_metastasis']

diabetes_model = tr.Model(target_cols, 'diabetes_mellitus')
diabetes_model.train(train_data)
test_data['predictions'] = diabetes_model.predict(test_data)

diabetes_model.accuracy(test_data, 'predictions')
```
